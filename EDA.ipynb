```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
# Load the dataset
file_path = "Housing.csv"
df = pd.read_csv(file_path)

# Display basic information and first few rows
df.info(), df.head()
# Summary statistics for numerical columns
summary_stats = df.describe()

# Checking for missing values
missing_values = df.isnull().sum()

# Checking unique values for categorical columns
unique_values = {col: df[col].unique() for col in df.select_dtypes(include=['object']).columns}
print(df.dtypes)
print(df.columns)
print(df.isnull().sum())
df = pd.get_dummies(df, drop_first=True)

# Checking correlation between numerical variables
correlation_matrix = df.corr()

summary_stats, missing_values, unique_values, correlation_matrix

# Boxplot: Price distribution by furnishing status
plt.figure(figsize=(8,6))
#df.rename(columns={'Furnishingstatus': 'furnishingstatus'}, inplace=True)
sns.boxplot(x=df['Furnishingstatus'], y=df['price'], palette="coolwarm")
plt.title("Price Distribution by Furnishing Status")
plt.show()
# Pair plot: Relationships between variables
sns.pairplot(df[['price', 'area', 'bedrooms', 'bathrooms', 'stories', 'parking']], diag_kind='kde', corner=True)
plt.show()

# Bar chart: Average price per number of bedrooms
plt.figure(figsize=(8,6))
sns.barplot(x=df['bedrooms'], y=df['price'], estimator=lambda x: sum(x)/len(x), palette="viridis")
plt.title("Average Price per Number of Bedrooms")
plt.xlabel("Number of Bedrooms")
plt.ylabel("Average Price")
plt.show()

# Regression plot: Predicting price based on area
plt.figure(figsize=(8,6))
sns.regplot(x=df['area'], y=df['price'], scatter_kws={'alpha':0.5}, line_kws={"color": "red"})
plt.xlabel("Area (sq ft)")
plt.ylabel("Price (in millions)")
plt.title("Regression Analysis: Price vs Area")
plt.show()

# Boxplot: Price distribution by furnishing status
plt.figure(figsize=(8,6))
sns.boxplot(x=df['Furnishingstatus'], y=df['price'], palette="coolwarm")
plt.title("Price Distribution by Furnishing Status")
plt.show()

# Pair plot: Relationships between variables
sns.pairplot(df[['price', 'area', 'bedrooms', 'bathrooms', 'stories', 'parking']], diag_kind='kde', corner=True)
plt.show()

# Bar chart: Average price per number of bedrooms
plt.figure(figsize=(8,6))
sns.barplot(x=df['bedrooms'], y=df['price'], estimator=lambda x: sum(x)/len(x), palette="viridis")
plt.title("Average Price per Number of Bedrooms")
plt.xlabel("Number of Bedrooms")
plt.ylabel("Average Price")
plt.show()

# Regression plot: Predicting price based on area
plt.figure(figsize=(8,6))
sns.regplot(x=df['area'], y=df['price'], scatter_kws={'alpha':0.5}, line_kws={"color": "red"})
plt.xlabel("Area (sq ft)")
plt.ylabel("Price (in millions)")
plt.title("Regression Analysis: Price vs Area")
plt.show()
```
