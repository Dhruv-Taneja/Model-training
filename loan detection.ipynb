```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Load the dataset
file_path = "loan_detection.csv"
df = pd.read_csv(file_path)

# Display basic information and first few rows
df.info(), df.head()


# Check class distribution
class_distribution = df["Loan_Status_label"].value_counts(normalize=True)

# Plot the distribution
plt.figure(figsize=(6, 4))
sns.barplot(x=class_distribution.index, y=class_distribution.values, palette="viridis")
plt.xticks([0, 1], ["Not Approved (0)", "Approved (1)"])
plt.ylabel("Proportion")
plt.title("Class Distribution of Loan_Status_label")
plt.show()

class_distribution
# Compute correlation matrix
correlation_matrix = df.corr()

# Focus on correlations with the target variable
target_corr = correlation_matrix["Loan_Status_label"].drop("Loan_Status_label").sort_values(ascending=False)

# Plot the top positive and negative correlations
plt.figure(figsize=(10, 6))
sns.barplot(x=target_corr.values[:10], y=target_corr.index[:10], palette="coolwarm", label="Top Positive")
sns.barplot(x=target_corr.values[-10:], y=target_corr.index[-10:], palette="coolwarm_r", label="Top Negative")
plt.axvline(0, color="black", linestyle="--", linewidth=1)
plt.xlabel("Correlation with Loan_Status_label")
plt.title("Top 10 Positive & Negative Correlations")
plt.show()

# Display the most positively and negatively correlated features
target_corr.head(10), target_corr.tail(10)

```
